```python
companies = ['Adani Ports','Asian Paints','Axis Bank','Bajaj Auto','Bajaj Finance','Bajaj Finserv','Bharat Petroleum','Britannia','Cipla',
            'Coal India','Dr Reddy','Eicher Motors','GAIL India','Grasim','HCL Technologies','HDFC Bank','HERO MotorCorp','Hindalco',
            'Hindustan Unilever','HDFC','ITC','ICICI Bank','Indian Oil','IndusInd Bank','Infosys','Kotak Bank','Larsen & Toubro',
            'Mahindra & Mahindra','Maruti Suzuki','NTPC','ONGC','Powergrid','Reliance Industries','State Bank of India','Sun Pharmaceutical','Tata Consultancy Services',
            'Tata Motors','Tata Steel','Tech Mahindra','Titan','UPL','UltraTech Cement','Vedanta','Wipro','Zee','Nestle', 'Shree Cement', 'JSW Steel']
```

#Predicting sentiments of news articles 

```python
import pandas as pd
from tqdm import tqdm

main_sentiment = pd.DataFrame()
main_score = pd.DataFrame()
i=0

for company in tqdm(companies):
  path = '/content/drive/My Drive/FYP/cleaned_2016-2021/'+company+'_news.csv'
  df = pd.read_csv(path)
  
  sentiment_df = pd.DataFrame()
  probs = []
  sentiments = []
  score=[]

  for text in df["cleaned_title"].to_list():
    sentence = flair.data.Sentence(text)
    sentiment_model.predict(sentence)
    probs.append(sentence.labels[0].score)      #Sentiment scores (0-1)
    sentiments.append(sentence.labels[0].value) #Sentiment labels (POSITIVE/NEGATIVE)

  for j in range(len(probs)):
    if sentiments[j] == 'NEGATIVE':
      score.append(-abs(probs[j]))
    else:
      score.append(probs[j])

  sentiment_df['published_at'] = df['published_at']
  sentiment_df['cleaned_title'] = df['cleaned_title']
  
    
  sentiment_df['probs'] = pd.Series(probs)
  sentiment_df['sentiments'] = pd.Series(sentiments)
  sentiment_df['score'] = pd.Series(score)
  sentiment_df['company_code'] = i
  sentiment_df['company_name'] = company

  sentiment_df['published_at']= pd.to_datetime(sentiment_df['published_at'])
  sentiment_df['year'] = pd.DatetimeIndex(sentiment_df['published_at']).year
  sentiment_df['month'] = pd.DatetimeIndex(sentiment_df['published_at']).month

  listcols = ['year','month','score','company_code']
  score_df = pd.DataFrame(sentiment_df.groupby(['year','month'])['score','company_code'].mean()).reset_index()
  score_df.column = listcols

  i=i+1

  main_sentiment = main_sentiment.append(sentiment_df, ignore_index=True)
  main_score = main_score.append(score_df, ignore_index=True)
#  break
```
